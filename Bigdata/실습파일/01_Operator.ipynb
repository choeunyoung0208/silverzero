### 자나깨나 오타조심  
에러문장을 잘 확인하자.  
Syntax Error : 내가 오타를 쳤구나.  
1) 쉼표와 점을 헷갈린다. .은 앞과 뒤를 연결짓는 기호다 ! --> .은 앞뒤의 공백이 없어요. / , 공백이 있다.  
2) :(콜론) ;(세미콜론) -> 모두 : 콜론이다.  
3) 여는괄호와 닫는괄호의 갯수를 잘 맞추자.  ((A, B, [C, (D)]))
