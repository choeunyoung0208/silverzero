# Colab으로 Python 시작하기

'#'은 글씨를 크게 만들고 싶을 때 사용하면 됨. '#'을 추가로 더 치면 글씨가 조금 작아짐.

```
print(1) #code를 실행시키는 방법 1 : ctrl + enter => 실행만 함.
```

```
1+5 #code를 실행시키는 방법 2 : shift + enter => 실행 후 아래 셀로 이동함.
```

ColabNotebook환경은 print를 굳이 쓰지 않아도 출력을 해줌.

```
1+5
4+6 #code를 실행시키는 방법 3 : alt + enter => 실행 후 아래 셀을 추가함.
```

4+6만 출력이 됨. 
=> ColabNotebook 환경에서 print를 하지 않았을 때도 출력이 되는 것은 기본적으로 하나의 셀 안에서 맨 마지막 줄에있는 명령만 출력이 됨.

```
1+2
3+4
5+7
```

ColabNotebook환경에서의 단축키 : ctrl + m 은 기본. (기본 단축키)
code cell에다 '#'을 추가해서 작성하는 것이 주석

```
# ctrl + m : 기본 -> 주석처리 : code cell 안에서 code에 포함시키고 싶지 않은 부분
```

code cell과 markdown cell을 서로 변형시키는 법

code cell -> markdown cell로 변환 : 기본 단축키 + m
markdown cell -> code cell로 변환 : 기본 단축키 + y

셀 삭제 : 기본 단축키 + d  
셀을 위에 추가 : 기본 단축키 + a  
셀을 아래에 추가 : 기본 단축키 + b
